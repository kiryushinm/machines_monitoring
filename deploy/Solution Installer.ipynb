# Machine State Monitoring — Solution Installer

This notebook deploys the **Machine Fleet State Monitoring** solution into the current
Microsoft Fabric workspace using the `fabric-launcher` library.

## Steps
1. **Deployment** — download source code, deploy all Fabric items
2. **Post-Deployment** — run configuration notebook

## 🚀 Deployment
Download source code and deploy Fabric items

```python
%pip install fabric-launcher --quiet
import notebookutils
notebookutils.session.restartPython()
```

```python
from fabric_launcher import FabricLauncher

launcher = FabricLauncher(
    notebookutils,
    api_root_url="https://api.fabric.microsoft.com"
)

launcher.download_and_deploy(
    repo_owner="makiryus_microsoft",
    repo_name="machines_state_monitoring",
    branch="main",
    workspace_folder="workspace",
    allow_non_empty_workspace=True,
    item_type_stages=[
        ["KQLDatabase", "Eventhouse"],
        ["Notebook", "Eventstream", "KQLQueryset"]
    ],
    validate_after_deployment=True,
    generate_report=True
)
```

## ✅ Post-Deployment Tasks

```python
import sempy.fabric as fabric
from fabric_launcher import move_item_to_folder

client = fabric.FabricRestClient()
workspace_id = fabric.get_workspace_id()
```

```python
# Move current notebook to the Install folder
move_item_to_folder(
    item_name=notebookutils.runtime.context["currentNotebookName"],
    item_type="Notebook",
    folder_name="Install",
    workspace_id=workspace_id,
    client=client
)
print("✅ Installer notebook moved to Install folder.")
```

```python
# Run PostDeploymentConfig notebook
result = launcher.run_notebook_synchronous(
    notebook_name="PostDeploymentConfig",
    parameters={},
    timeout_seconds=3600
)
print("✅ Post-deployment configuration complete!")
```
